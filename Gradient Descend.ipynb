```python
import numpy as ny
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    m_curr = 0;b_curr = 0
    #reach global minima
    #so there are this (x,y) points. When plotted their intercept_vs_slope_vs_mse graph,what we want is minima of that graph
    #that point where we get minima is our real m,c
    #we got our regression line. Not by sklearn, not by m = mx+c old school but by gradient descent
    #few things we need to take care as below
    iterations = 10000
    n = len(x)
    learning_rate = 0.01
    #itereation and learning rate are paramaters that should be initalized with some value
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        
        #cost function or mse here below. In sum function make list of nos. and now sum them
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        
        md = (-2/n)*sum(x*(y - y_predicted))
        bd = (-2/n)*sum(y-y_predicted)
        
        m_curr = m_curr - learning_rate*md
        b_curr = b_curr - learning_rate*bd
        
        #print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
        
        plt.plot(x,m_curr*x + b_curr, color = 'c',zorder=1)
    
    
    plt.plot(x,m_curr*x + b_curr, color = 'b', zorder=2)
    plt.scatter(x,y, c = 'r', zorder=3,marker = 'o')
    
```

```python
x = ny.array([1,2,3,4,5])
y = ny.array([5,7,9,11,13])

gradient_descent(x,y)
```
